```python
import matplotlib.pyplot as plt

fil = open("kolsaas.csv")
data = fil.read()
```

```python
linjer = data.split("\n")
tur = []
for linje in linjer:
    splittet_linje = linje.split(",")
    tur.append(splittet_linje)
tur
```

```python
yakse = []
xakse = []

for linjer in tur[1:]:
    yakse.append(linjer[3])
    xakse.append(linjer[0])
```

```python
plt.plot(xakse, yakse)
plt.show()
```
